```python
import pandas as pd
import numpy as np

# Load revenue_geo_segment.csv
df = pd.read_csv('revenue_geo_segment.csv')

# Get data from the first and second columns (indexes 0 and 1)
symbol = df.iloc[:, 0]
data = df.iloc[:, 1]

# Create empty lists to store the extracted data
symbols = []
dates = []
regions = []
revenues = []

# Iterate over the data and extract information
for sym, item in zip(symbol, data):
    if pd.isnull(item):
        # Append None values if the item is NaN
        symbols.append(sym)
        dates.append(None)
        regions.append(None)
        revenues.append(None)
    else:
        try:
            # Try to parse the item as a dictionary
            item_dict = eval(item)
            # Extract date from the dictionary key
            date = list(item_dict.keys())[0]
            # Iterate over the regions and revenues and append them to the respective lists
            for region, revenue in item_dict[date].items():
                symbols.append(sym)
                dates.append(date)
                regions.append(region)
                revenues.append(revenue)
        except Exception as e:
            # Handle any exceptions (e.g., invalid data format)
            print(f"Error processing item: {item}. Error: {e}")
            symbols.append(sym)
            dates.append(None)
            regions.append(None)
            revenues.append(None)

# Create a DataFrame from the extracted data
revenue_geo_segment_df = pd.DataFrame({
    'symbol': symbols,
    'date': dates,
    'region': regions,
    'revenue': revenues
})

# Display the DataFrame
revenue_geo_segment_df
```

```python
import pandas as pd

# Load revenue_product_segment.csv
df = pd.read_csv('revenue_product_segment.csv')

# Create empty lists to store data
symbols = []
dates = []
segments = []
revenues = []

# Iterate over each row in the DataFrame
for _, row in df.iterrows():
    symbol = row['symbol']
    data = row['revenue_product_segment']
    if pd.notna(data):
        for date, segment_revenue in eval(data).items():
            for segment, revenue in segment_revenue.items():
                # Append symbol, date, segment, and revenue to the lists
                symbols.append(symbol)
                dates.append(date)
                segments.append(segment)
                revenues.append(revenue)
    else:
        # If data is NaN, append None values
        symbols.append(None)
        dates.append(None)
        segments.append(None)
        revenues.append(None)

# Create a DataFrame from the lists
revenue_product_segment_df = pd.DataFrame({
    'symbol': symbols,
    'date': dates,
    'segment': segments,
    'revenue': revenues
})

# Display the DataFrame
revenue_product_segment_df
```

```python
import psycopg2
from psycopg2 import sql

# Establish a connection to the PostgreSQL database
conn = psycopg2.connect(
    dbname="stockStock",
    user="postgres",
    password="123",
    host="localhost"
)
cur = conn.cursor()

# Create the symbols_list table
cur.execute("""
CREATE TABLE IF NOT EXISTS symbols_list (
    symbol VARCHAR(10) PRIMARY KEY
);
""")

# Create the balance_sheet table
cur.execute("""
CREATE TABLE IF NOT EXISTS balance_sheet (
    id SERIAL PRIMARY KEY,
    date DATE NOT NULL,
    symbol VARCHAR(10) NOT NULL,
    totaldebt FLOAT,
    CONSTRAINT symbol_fk FOREIGN KEY (symbol) REFERENCES symbols_list (symbol) ON DELETE NO ACTION ON UPDATE CASCADE
);
""")

# Create the cashflow_statement table
cur.execute("""
CREATE TABLE IF NOT EXISTS cashflow_statement (
    id SERIAL PRIMARY KEY,
    date DATE NOT NULL,
    symbol VARCHAR(10) NOT NULL,
    freecashflow FLOAT,
    CONSTRAINT symbol_fk FOREIGN KEY (symbol) REFERENCES symbols_list (symbol) ON DELETE NO ACTION ON UPDATE CASCADE
);
""")

# Create the company_profiles table
cur.execute("""
CREATE TABLE IF NOT EXISTS company_profiles (
    id SERIAL PRIMARY KEY,
    symbol VARCHAR(10) NOT NULL,
    companyname VARCHAR(255),
    industry VARCHAR(255),
    sector VARCHAR(255),
    beta FLOAT, 
    mktcap FLOAT, 
    CONSTRAINT symbol_fk FOREIGN KEY (symbol) REFERENCES symbols_list (symbol) ON DELETE NO ACTION ON UPDATE CASCADE
);
""")

# Create the earning_call_transcript table
cur.execute("""
CREATE TABLE IF NOT EXISTS earning_call_transcript (
    id SERIAL PRIMARY KEY,
    date DATE NOT NULL,
    symbol VARCHAR(10) NOT NULL,
    transcript TEXT,
    CONSTRAINT symbol_fk FOREIGN KEY (symbol) REFERENCES symbols_list (symbol) ON DELETE NO ACTION ON UPDATE CASCADE
);
""")

# Create the financial_ratios table
cur.execute("""
CREATE TABLE IF NOT EXISTS financial_ratios (
    id SERIAL PRIMARY KEY,
    date DATE NOT NULL,
    symbol VARCHAR(10) NOT NULL,
    debtRatio FLOAT, 
    debtEquityRatio FLOAT,
    cashFlowToDebtRatio FLOAT,
    priceToBookRatio FLOAT,
    priceToSalesRatio FLOAT,
    priceEarningsRatio FLOAT,
    priceEarningsToGrowthRatio FLOAT,
    receivablesTurnover FLOAT,
    payablesTurnover FLOAT,
    assetTurnover FLOAT,
    currentRatio FLOAT,
    quickRatio FLOAT,
    cashRatio FLOAT,
    returnOnEquity FLOAT,
    grossProfitMargin FLOAT,
    netProfitMargin FLOAT,
    dividendPayoutRatio FLOAT,        
    CONSTRAINT symbol_fk FOREIGN KEY (symbol) REFERENCES symbols_list (symbol) ON DELETE NO ACTION ON UPDATE CASCADE
);
""")

# Create the historical_price table
cur.execute("""
CREATE TABLE IF NOT EXISTS historical_price (
    id SERIAL PRIMARY KEY,
    symbol VARCHAR(10) NOT NULL,
    date DATE NOT NULL,
    open FLOAT,
    high FLOAT,
    low FLOAT,
    close FLOAT,
    volume BIGINT,
    CONSTRAINT symbol_fk FOREIGN KEY (symbol) REFERENCES symbols_list (symbol) ON DELETE NO ACTION ON UPDATE CASCADE
);
""")

# Create the income_statement table
cur.execute("""
CREATE TABLE IF NOT EXISTS income_statement (
    id SERIAL PRIMARY KEY,
    date DATE NOT NULL,
    symbol VARCHAR(10) NOT NULL,
    revenue FLOAT,
    netIncome FLOAT,
    CONSTRAINT symbol_fk FOREIGN KEY (symbol) REFERENCES symbols_list (symbol) ON DELETE NO ACTION ON UPDATE CASCADE
);
""")

# Create the insider_trading_statistic table
cur.execute("""
CREATE TABLE IF NOT EXISTS insider_trading_statistic (
    id SERIAL PRIMARY KEY,
    date DATE NOT NULL,
    symbol VARCHAR(10) NOT NULL,
    total_insider_shares_traded FLOAT,
    total_insider_shares_held FLOAT,
    percent_insider_shares_held FLOAT,
    CONSTRAINT symbol_fk FOREIGN KEY (symbol) REFERENCES symbols_list (symbol) ON DELETE NO ACTION ON UPDATE CASCADE
);
""")

# Create the insider_trading table
cur.execute("""
CREATE TABLE IF NOT EXISTS insider_trading (
    id SERIAL PRIMARY KEY,
    date DATE NOT NULL,
    symbol VARCHAR(10) NOT NULL,
    insider_name VARCHAR(255),
    relation VARCHAR(255),
    last_date FLOAT,
    transaction_type VARCHAR(255),
    own_type VARCHAR(255),
    shares_traded FLOAT,
    last_price FLOAT,
    shares_held FLOAT,
    CONSTRAINT symbol_fk FOREIGN KEY (symbol) REFERENCES symbols_list (symbol) ON DELETE NO ACTION ON UPDATE CASCADE
);
""")

# Create the revenue_geo_segment table
cur.execute("""
CREATE TABLE IF NOT EXISTS revenue_geo_segment (
    id SERIAL PRIMARY KEY,
    date DATE NOT NULL,
    symbol VARCHAR(10) NOT NULL,
    country VARCHAR(255),
    revenue FLOAT,
    CONSTRAINT symbol_fk FOREIGN KEY (symbol) REFERENCES symbols_list (symbol) ON DELETE NO ACTION ON UPDATE CASCADE
);
""")

# Create the revenue_product_segment table
cur.execute("""
CREATE TABLE IF NOT EXISTS revenue_product_segment (
    id SERIAL PRIMARY KEY,
    date DATE NOT NULL,
    symbol VARCHAR(10) NOT NULL,
    product_category VARCHAR(255),
    revenue FLOAT,
    CONSTRAINT symbol_fk FOREIGN KEY (symbol) REFERENCES symbols_list (symbol) ON DELETE NO ACTION ON UPDATE CASCADE
);
""")

# Commit changes and close the connection
conn.commit()
conn.close()
```
